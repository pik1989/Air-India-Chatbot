## RAG Chatbot

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/Aiesl Employees service regulation.pdf")
documents = loader.load()

context = "\n\n".join(doc.page_content for doc in documents)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    You are an intelligent assistant trained to understand and provide information based
    on the context provided.

    Context: {context}
    Question: {question}

    Your responses should directly address the question using the information from the context,
    while being clear and concise.
    """
)
```

```python
formatted_prompt = prompt.format(
    context=context,
    question="What is the maximum number of days of Full Pay Leave an employee can accumulate"
)
```

```python
result = chat.invoke(formatted_prompt)
```

```python
print(result.content)
```

```python
for m in chat.stream(formatted_prompt):
  print(m.content, end="", flush=True)
```

## Directory Based Chatbot

```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader

loader = DirectoryLoader(
    path="/content/test",
    glob="*/.pdf",
    loader_cls=PyPDFLoader
)

documents = loader.load()
```

```python
context = "\n\n".join(doc.page_content for doc in documents)
```

```python
formatted_prompt = prompt.format(
    context=context,
    question="Air India Employee study leaves, write about this topic in 2 lines"
)
```

```python
for m in chat.stream(formatted_prompt):
  print(m.content, end="", flush=True)
```
